```python
listă_de_frecvențe = [0] * 11
```

```python
note = [1]*10 + [2]*15 + [3]*30 + [4]*25 + [5]*20 + [6]*30 + [7]*40 + [8]*60 + [9]*75 + [10]*50

for notă in note:
	listă_de_frecvențe[notă] += 1

print(listă_de_frecvențe)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

grades = [1]*10 + [2]*15 + [3]*30 + [4]*25 + [5]*20 + [6]*30 + [7]*40 + [8]*60 + [9]*75 + [10]*50

mu, std = np.mean(grades), np.std(grades)
n = len(grades)

fig, ax = plt.subplots()

# histogram
ax.hist(grades, bins=10, range=(1, 11), color='orange', edgecolor='black')


# tick centered on each bin, all 10 grades showing
ax.set_xticks([i + 0.5 for i in range(1, 11)])
ax.set_xticklabels(range(1, 11))

ax.set_xlabel('Note')
ax.set_ylabel('Frecvență')
ax.set_title('Distribuția notelor')
ax.legend()
plt.show()
```

```python
masini = ["BMW albastru", "mini cooper negru", "volkswagen rosu", "fiat maro"]
```

```python
class Node:
	def __init__(self, info: str):
		self.info = info
		self.next = None
```

```python
masini = ["BMW albastru", "mini cooper negru", "volkswagen rosu", "fiat maro"]

head = Node(masini[0])
current = head

for i in range(1, len(masini)):
	new_node = Node(masini[i])
	current.next = new_node
	current = new_node

current = head
while current is not None:
	print(current.info)
	current = current.next
```
