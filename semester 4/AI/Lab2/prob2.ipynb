```python
import os
from PIL import Image

folder = "images"
images = []

for file in os.listdir(folder):
    path = os.path.join(folder, file)
    img_file = Image.open(path)
    img_file = img_file.resize((128, 128))
    img_file.filename = path
    images.append(img_file)

plt.figure(figsize=(10,10))

for index, image in enumerate(images):
    plt.subplot(3, 4, index + 1)
    plt.imshow(image)
    plt.title(image.filename.split('/')[-1])
    plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(10,10))

for index, image in enumerate(images):
    # https://www.geeksforgeeks.org/python/python-pillow-tutorial/
    plt.subplot(3, 4, index + 1)
    plt.imshow(image.convert("L"), cmap='gray') # 8-bit pixels, Grayscale
    plt.title(image.filename.split('/')[-1])
    plt.axis('off')

plt.show()
```

```python
from PIL import ImageFilter

img_prob3 = Image.open(r'images/LeCun.jpg')

plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.imshow(img_prob3)
plt.axis('off')
plt.title('Defore blurr')

plt.subplot(1, 2, 2)
plt.imshow(img_prob3.filter(ImageFilter.GaussianBlur(10)))
plt.axis('off')
plt.title('After blurr')

plt.show()
```

```python
img = Image.open(r"images/Karpaty.jpg")
edges = img.convert("L").filter(ImageFilter.FIND_EDGES)

plt.figure(figsize=(10,10))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off')
plt.title('Defore blurr')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('After blurr')

plt.show()
```
