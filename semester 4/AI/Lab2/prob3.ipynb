```python
from nltk.tokenize import word_tokenize, sent_tokenize

with open("data/texts.txt", "r", encoding="utf-8") as f:
    text = f.read()

sentences = sent_tokenize(text)

print("Number of sentences:", sentences.__len__())
```

```python
words = [w for w in word_tokenize(text) if w.isalpha()]
words.__len__()
```

```python
import numpy as np
np.unique(words).__len__()
```

```python
longest_word = max(words, key=len)
print("Longest word: ", longest_word)
print("Shortest word: ", min(words, key=len))
```

```python
import unicodedata

def remove_diacritics(text):
    # Normalize text to NFD (decomposed form)
    nfkd_form = unicodedata.normalize('NFD', text)
    # Keep only characters that are not diacritical marks
    return ''.join([c for c in nfkd_form if not unicodedata.combining(c)])

print(remove_diacritics(text))
```

```python
from nltk.corpus import wordnet

synonyms = wordnet.synsets("Confirm") # get all the meaning of the word
print("Synonyms of the longest word:")
for syn in set([lemma.name() for syn in synonyms for lemma in syn.lemmas()]):
    print(syn)
```
