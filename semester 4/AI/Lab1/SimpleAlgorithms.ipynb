## Problem 1

```python
# Complexity: O(N)
def last_alphabetic_word():
    text = input().strip()
    last = ''
    for word in text.split():
        if last < word:
            last = word

    return last
```

```python
# ChatGPT version
def ai_last_alphabetic_word():
    text = input().strip()
    words = text.split(" ")
    last_word = max(words)
    return last_word
```

## Problem 2

First method is with Herons Algorithm, even if *Quake's* (from the game) formula is faster, but for python to impractical. The second method uses binary search.

```python
# Complexity: O(log(N))
def sqrt_Heron(number: float) -> float:
    if number == 0:
        return 0

    if number < 0:
        raise ValueError("sqrt_Heron(s) is not defined for negative numbers.")

    precision = 0.01 # 1e-7 10**-7
    difference = 1
    guess: float = number / 2

    while difference > precision:
        aux = 0.5 * (guess + number / guess)
        difference = abs(guess - aux)
        guess = aux

    return guess

def euclidean_distance(point1, point2, function):
    return function(
        (point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2,
    )

print( (1,5), (4,1), sqrt_Heron )
```

```python
# Complexity: O(log(N))
def sqrt_binarySearch(number: float) -> float:
    start, end = 0, number
    precision = 0.001

    while end - start > precision:
        mid = (start + end) / 2
        aux = mid ** 2
        if number - precision < aux < number + precision:
            return mid

        if aux < number:
            start = mid + precision
        else:
            end = mid - precision

    return (start + end) / 2

print( (1,5), (4,1), sqrt_binarySearch)
```

## Problem 3
First method with map for sparse vectors

```python
# Complexity: O(N+M) where N, M are the numbers !0 from the vectors
def scalar_product() -> float:
    vecA: dict[int, int] = {
        i: int(v)
        for i, v in enumerate(map(int, input().split())) if v != 0
    }

    vecB: dict[int, int] = {
        i: int(v)
        for i, v in enumerate(map(int, input().split())) if v != 0
    }

    product: float = 0.0
    for key in vecA.keys():
        product += vecA[key] * vecB[key]
        del vecB[key]

    for key in vecB.keys():
        product += vecB[key] * vecA[key]

    return product
```

```python
def scalar_product_vector(vecA: list[int], vecB:list[int]) -> float:
    product = 0.0
    for a, b in zip(vecA, vecB):
        product += a * b

    return product
```

```python
# ChatGPT version
def ai_scalar_product():
    v1 = list(map(float, input().split()))
    v2 = list(map(float, input().split()))

    dot_product = 0.0

    # compute dot product
    for a, b in zip(v1, v2):
        if a != 0 and b != 0:
            dot_product += a * b

    print(dot_product)
```

## Problem 4

```python

# Complexity: O(n)
def appears_once():
    text = input()
    word_count = {}

    for word in text.split():
        word_count[word] = word_count.get(word, 0) + 1

    sol:list[str] = []
    for (word, count) in sorted(word_count.items(), key=lambda item: item[1]):
        if count > 1:
            return sol
        sol.append(word)

    return sol
```

```python
# ChatGPT version
def ai_appears_once():
    text = input()
    words = text.split()
    frequency = {}

    # count occurrences
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    # collect words that appear once
    result = []
    for word in words:
        if frequency[word] == 1:
            result.append(word)

    return result
```

## Problem 5

```python
# Complexity: O(n)
def find_duplicate(numbers: list[int]) -> int:
    return sum(numbers) - len(numbers) * (len(numbers) - 1) // 2
```

```python
# ChatGPT version
def ai_find_duplicate(nums):
    seen = set()

    for number in nums:
        if number in seen:
            return number
        seen.add(number)
```

## Problem 6

```python
# Complexity: O(n)
def majority_element(numbers: list[int]) -> int:
    length = len(numbers)
    first = second = 0
    first_number = -1
    mapping: dict[int, int] = {}

    for (index, number) in enumerate(numbers):
        mapping[number] = mapping.get(number, 0) + 1
        if first < mapping[number]:
            first = mapping[number]
            first_number = number
        elif second < mapping[number]:
            second = mapping[number]

        if first > second + (length - index):
            return first_number

    return first_number
```

```python
# ChatGPT version
def ai_majority_element(nums):
    candidate = None
    count = 0

    # Boyer-Moore Voting Algorithm
    for number in nums:
        if count == 0:
            candidate = number
        if number == candidate:
            count += 1
        else:
            count -= 1

    return candidate
```

## Problem 7

```python
# O(n Log(n))
def k_biggest_number(nums: list[int], k) -> int:
    numbers = list(set(nums))       # WC Th(n^2) - if the hash is bad, on average O(n)
    numbers.sort(reverse=True)      # O(n Log(n))
    return numbers[k - 1]
```

```python
# ChatGPT version
def ai_k_biggest_number(nums, k):
    sorted_nums = sorted(nums, reverse=True)
    return sorted_nums[k - 1]
```

## Problem 8

```python
# Complexity: O(Log(n))
def get_binary(n: int) -> list[int]:
    bits = []
    while True:
        bits.append(n & 1)
        n //= 2
        if n == 0:
            break

    bits.reverse()
    return bits

# Complexity: O(log n!)
def generate_binary_numbers(n: int) -> list[list[int]]:
    binary_numbers = []
    for i in range(n):
        binary_numbers.append(get_binary(i))
    return binary_numbers
```

```python
# ChatGPT version
def ai_generate_binary_numbers(n):
    result = []
    for i in range(1, n + 1):
        result.append(bin(i)[2:])
    return result
```
